# Part 11: Clonal Evolution

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

DATA_PATH = Path("41586_2022_5154_MOESM3_ESM.xlsx")
FIG_DIR = Path("figures")
FIG_DIR.mkdir(exist_ok=True)
s1 = pd.read_excel(DATA_PATH, sheet_name="TableS1_clinical_data")
s11 = pd.read_excel(DATA_PATH, sheet_name="TableS11_evolution")
sa_ids_T = {
    sid + "-T"
    for sid in set(s1[s1["Country"] == "South Africa"]["Sample_id"].astype(str))
}
```

```python
s11["id_T"] = s11["Sample"].astype(str)
s11_sa = s11[s11["id_T"].isin(sa_ids_T)].copy()
print("n SA matched:", len(s11_sa))
print("Nodes median/mean:", s11_sa["Nodes"].median(), s11_sa["Nodes"].mean())
print(
    "Subclonal SSM median/mean:",
    s11_sa["subclonal_num_ssms"].median(),
    s11_sa["subclonal_num_ssms"].mean(),
)
display(s11_sa["Pattern"].value_counts())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
max_nodes = int(s11_sa["Nodes"].max())
axes[0].hist(
    s11_sa["Nodes"], bins=range(1, max_nodes + 2), color="#8e44ad", edgecolor="white"
)
axes[0].set_title("Clonal Complexity — SA Cohort")
axes[1].hist(
    s11_sa["subclonal_num_ssms"].dropna(), bins=30, color="#16a085", edgecolor="white"
)
axes[1].set_title("Subclonal Mutation Burden — SA Cohort")
fig.tight_layout()
fig.savefig(FIG_DIR / "fig13_clonal_evolution.png", dpi=200)
plt.close(fig)
```
