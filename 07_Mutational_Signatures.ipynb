# Part 7: Mutational Signatures

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

DATA_PATH = Path("41586_2022_5154_MOESM3_ESM.xlsx")
FIG_DIR = Path("figures")
FIG_DIR.mkdir(exist_ok=True)
s1 = pd.read_excel(DATA_PATH, sheet_name="TableS1_clinical_data")
s9 = pd.read_excel(DATA_PATH, sheet_name="TableS9_enrichment")
sa_ids = set(s1[s1["Country"] == "South Africa"]["Sample_id"].astype(str))
```

```python
s9["id_str"] = s9["Sample"].astype(str)
s9_sa = s9[s9["id_str"].isin(sa_ids)].copy()
sig_cols = [
    "SBS_cosmic_top_enriched",
    "DBS_cosmic_top_enriched",
    "ID_cosmic_top_enriched",
    "CN_top_enriched",
    "SV_top_enriched",
]
for col in sig_cols:
    print("\n", col)
    display(s9_sa[col].value_counts().head(8))
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes_flat = axes.flatten()
for i, col in enumerate(sig_cols):
    vc = s9_sa[col].value_counts().head(8)
    axes_flat[i].bar(vc.index, vc.values, color=sns.color_palette("Set2", len(vc)))
    axes_flat[i].set_title(col.replace("_top_enriched", "").replace("_cosmic", ""))
    axes_flat[i].tick_params(axis="x", rotation=45)
axes_flat[-1].set_visible(False)
fig.tight_layout()
fig.savefig(FIG_DIR / "fig9_mutational_signatures.png", dpi=200)
plt.close(fig)
```
