# Part 5: Molecular Subtype Distribution
Distribution: supermarkets, specialist shops, molecular type.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

DATA_PATH = Path("41586_2022_5154_MOESM3_ESM.xlsx")
FIG_DIR = Path("figures")
FIG_DIR.mkdir(exist_ok=True)
s1 = pd.read_excel(DATA_PATH, sheet_name="TableS1_clinical_data")
s5a = pd.read_excel(DATA_PATH, sheet_name="TableS5_TCGA")
sa = s1[s1["Country"] == "South Africa"].copy()
sa_ids = set(sa["Sample_id"].astype(str))
```

```python
s5a["id_str"] = s5a["id"].astype(str)
s5a_sa = s5a[s5a["id_str"].isin(sa_ids)].copy()
gms_counts = s5a_sa["subtype"].value_counts().sort_index()
s5a_sa["ETS_binary"] = (s5a_sa["ETS_fusion"] != "neg").astype(int)
ets_pct = 100 * s5a_sa["ETS_binary"].mean()
type_counts = s5a_sa["Type"].value_counts()
display(gms_counts)
print(f"ETS fusion prevalence: {ets_pct:.1f}%")
display(type_counts)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
gms_sa_pct = gms_counts / gms_counts.sum() * 100
axes[0].bar([f"GMS-{s}" for s in gms_sa_pct.index], gms_sa_pct.values)
axes[0].set_title(f"GMS Subtypes — SA (n={len(s5a_sa)})")
axes[0].set_ylabel("Percentage (%)")
type_pct = type_counts / type_counts.sum() * 100
axes[1].barh(type_pct.index[::-1], type_pct.values[::-1])
axes[1].set_title(f"Molecular Type — SA (n={len(s5a_sa)})")
axes[1].set_xlabel("Percentage (%)")
fig.tight_layout()
fig.savefig(FIG_DIR / "fig7_gms_subtypes.png", dpi=200)
plt.close(fig)
```
